# 2. Creating Functions and For loops for price_range and Busiest day

```python
# Define a function

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return np.nan
    
# If the value within the “prices” column within the given row is less than or equal to 5, then return the string “Low-range product.”

# Or else, if the value within the “prices” column within the given row is greater than 5 and less than or equal to 15, then return the string “Mid-range product.”

# Or else, if the value of the “prices” column within the given row is greater than 15, then return the string “High-range product.”

# Or else, return the string “Not enough data.”
```

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

# 3. Creating columns and for loops for Busiest Days and Busiest Hours

```python
# Frequency check for orders_day_of_week column
df_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Creating Column to show busiest vs slowest days

results = []

for value in df_merged['orders_day_of_week']:
    if value ==  0 or value == 1:
        results.append("Busiest Days")
    elif value == 4 or value == 3:
        results.append("Slowest Days")
    else:
        results.append("Regularly Busy")
```

```python
df_merged['busiest days'] = results
```

```python
# Frequency check for new busiest days column
df_merged['busiest days'].value_counts(dropna = False)
```

```python
# Frequency check for hour_of_day_ordered column
df_merged['hour_of_day_ordered'].value_counts(dropna = False)
```

```python
# Creating a for loop with data dictionary for hour_of_day_orderd 

hours = []

hours_counts = {
    (0, 2, 3, 4, 5, 6): "Fewest Orders",
    (10, 11, 12, 13, 14, 15): "Most Orders"
}

for value in df_merged['hour_of_day_ordered']:
    for keys, label in hours_counts.items():
        if value in keys:
            hours.append(label)
            break
    else:
        hours.append("Average orders")
       
```

```python
df_merged['Busiest hours'] = hours
```

```python
# Frequency Check for busiest hours column
df_merged['Busiest hours'].value_counts(dropna = False)
```
