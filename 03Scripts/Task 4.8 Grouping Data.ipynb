# 2. Creating columns and flags

```python
# average number of orders by department_id

ords_prods_merged.groupby('department_id').agg({'order_number' : ['mean']})
```

# The output for the entire dataframe includes all of the departements, also the averages for the departements we looked in in the subset have changed slightly.

```python
# New column for max_orders
ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merged.head(10)
```

```python
# Creating loyalty flags
# loyal Customers
ords_prods_merged.loc[ords_prods_merged['max_order'] >40, 'loyalty_flag'] = 'Loyal Customer'
```

```python
# Regular Customer
ords_prods_merged.loc[(ords_prods_merged['max_order']
                       <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
```

```python
# New Customer
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

```python
# Check
ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check
ords_prods_merged[['user_id', 'loyalty_flag', 'order_number']].head(20)
```

```python
# price statistics by loyalty flag
ords_prods_merged.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

# New & Regular customers seem to buy higher priced items, while loyal customers buy lower priced items. 

```python
# Creating new column for max_prices 
ords_prods_merged['avg_prices'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Check
ords_prods_merged.head(10)
```

```python
# creating flags based on customer spending
# High Spender

ords_prods_merged.loc[ords_prods_merged['avg_prices'] >=10, 'spending_flag'] = 'High Spender'
```

```python
# Low Spender

ords_prods_merged.loc[ords_prods_merged['avg_prices'] <10, 'spending_flag'] = 'Low Spender'
```

```python
# Check
ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

```python
# Check
ords_prods_merged[['user_id', 'spending_flag', 'avg_prices']].head(100)
```

```python
# Create order frequency column by median
ords_prods_merged['order_frequency'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merged.head(10)
```

```python
# create order frequency flags
# Frequent customers

ords_prods_merged.loc[ords_prods_merged['order_frequency'] <= 10, 'frequency_flag'] = 'Frequent Customers'
```

```python
# Regular Customers

ords_prods_merged.loc[(ords_prods_merged['order_frequency'] >=10) &
                      (ords_prods_merged['order_frequency'] <=20), 'frequency_flag'] = 'Regular Customer'
```

```python
# Non-Frequent Customers

ords_prods_merged.loc[ords_prods_merged['order_frequency'] >20, 'frequency_flag'] = 'Non-Frequent Customer'
```

```python
#Check
ords_prods_merged['frequency_flag'].value_counts(dropna = False)
```

```python
#Check
ords_prods_merged[['order_id', 'frequency_flag', 'order_frequency']].head(100)
```
